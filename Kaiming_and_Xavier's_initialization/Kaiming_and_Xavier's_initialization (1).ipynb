```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import layers, models, initializers, regularizers
import matplotlib.pyplot as plt
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
y_train, y_test = to_categorical(y_train), to_categorical(y_test)
def create_model(activation="relu", regularizer=None, initializer=None):
    model = models.Sequential([
        layers.Flatten(input_shape=(32, 32, 3)),
        layers.Dense(512, activation=activation, kernel_regularizer=regularizer,
                     kernel_initializer=initializer),
        layers.Dense(256, activation=activation, kernel_regularizer=regularizer,
                     kernel_initializer=initializer),
        layers.Dense(128, activation=activation, kernel_regularizer=regularizer,
                     kernel_initializer=initializer),
        layers.Dense(64, activation=activation, kernel_regularizer=regularizer,
                     kernel_initializer=initializer),
        layers.Dense(32, activation=activation, kernel_regularizer=regularizer,
                     kernel_initializer=initializer),
        layers.Dense(10, activation='softmax')
    ])
    return model
xavier_initializer = initializers.glorot_normal()
kaiming_initializer = initializers.he_normal()
regularizer = regularizers.l2(0.001)
xavier_model = create_model(activation='sigmoid', regularizer=regularizer, initializer=xavier_initializer)
xavier_model.compile(optimizer='adam',
                     loss='categorical_crossentropy',
                     metrics=['accuracy'])
kaiming_model = create_model(activation='relu', regularizer=regularizer, initializer=kaiming_initializer)
kaiming_model.compile(optimizer='adam',
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])
print("Xavier Model Summary:")
xavier_model.summary()
print("\nKaiming Model Summary:")
kaiming_model.summary()
print("\nTraining Xavier Model...")
xavier_history = xavier_model.fit(x_train, y_train, epochs=25,
                                  batch_size=64,
                                  validation_data=(x_test, y_test),
                                  verbose=1)
print("\nTraining Kaiming Model...")
kaiming_history = kaiming_model.fit(x_train, y_train, epochs=25,
                                    batch_size=64,
                                    validation_data=(x_test, y_test),
                                    verbose=1)
fig, ax = plt.subplots(1, 2, figsize=(16, 8))
ax[0].plot(xavier_history.history['loss'], label='train loss')
ax[0].plot(xavier_history.history['val_loss'], label='val loss')
ax[0].set_title('Xavier Model Loss')
ax[0].set_ylabel('Loss')
ax[0].set_xlabel('Epoch')
ax[0].legend(loc='upper left')
ax[1].plot(xavier_history.history['accuracy'], label='train accuracy')
ax[1].plot(xavier_history.history['val_accuracy'], label='val accuracy')
ax[1].set_title('Xavier Model Accuracy')
ax[1].set_ylabel('Accuracy')
ax[1].set_xlabel('Epoch')
ax[1].legend(loc='upper left')
plt.tight_layout()
plt.show()
fig, ax = plt.subplots(1, 2, figsize=(16, 8))
ax[0].plot(kaiming_history.history['loss'], label='train loss')
ax[0].plot(kaiming_history.history['val_loss'], label='val loss')
ax[0].set_title('Kaiming Model Loss')
ax[0].set_ylabel('Loss')
ax[0].set_xlabel('Epoch')
ax[0].legend(loc='upper left')
ax[1].plot(kaiming_history.history['accuracy'], label='train accuracy')
ax[1].plot(kaiming_history.history['val_accuracy'], label='val accuracy')
ax[1].set_title('Kaiming Model Accuracy')
ax[1].set_ylabel('Accuracy')
ax[1].set_xlabel('Epoch')
ax[1].legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
loss_xavier, accuracy_xavier = xavier_model.evaluate(x_test, y_test, verbose=0)
loss_kaiming, accuracy_kaiming = kaiming_model.evaluate(x_test, y_test, verbose=0)

print(f"Xavier Model Test Accuracy: {accuracy_xavier:.4f}")
print(f"Kaiming Model Test Accuracy: {accuracy_kaiming:.4f}")
```
